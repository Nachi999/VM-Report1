```python
######################### Category TABLE ###############################

category_table = pd.pivot_table(data=join_RTTA, 
                        index=['Category'], 
                        values='Title',
                        aggfunc='count')
category_table = category_table.reset_index()
category_table = category_table.rename(columns = {"Title": "No. of Vulnerabilities"})

severity_template = pd.read_excel("C:\\Users\\nthakur\\Downloads\\Python nachi\\Extract Crown Jewels reports\\Template\\Severity Template.xlsx")


######################### Severity TABLE 2 ###############################

Severity_table = pd.pivot_table(data=join_RTTA, 
                        index=['Severity'], 
                        values='Title',
                        aggfunc='count')
Severity_table = Severity_table.reset_index()
Severity_table = Severity_table.rename(columns = {"Title": "No. of Vulnerabilities"})
Severity_table = Severity_table.sort_values(by="Severity",ascending=False)

######################### Severity TABLE 3 ###############################

asset_table = pd.pivot_table(data=join_RTTA, 
                        index=['Host Name'], 
                        values='Title',
                        aggfunc='count')
asset_table = asset_table.reset_index()
asset_table = asset_table.rename(columns = {"Title": "No. of Vulnerabilities"})
asset_table = asset_table.sort_values(by="No. of Vulnerabilities",ascending=False)

############################WRITTING#######################

with pd.ExcelWriter(path,engine="xlsxwriter") as writer:
    Severity_table.to_excel(writer,sheet_name = "Severity", startrow = 2,index=False)     
    asset_table.to_excel(writer,sheet_name = "Assets", startrow = 2,index=False)
    category_table.to_excel(writer,sheet_name = "Category", startrow = 2,index=False) 
    join_RTTA.to_excel(writer,sheet_name = sheet_name, startrow = 2, index=False)
    
  ###########FORMATTING DETAILED REPORT - TABLE #################### 

    # Get the xlsxwriter workbook and worksheet objects

    workbook  = writer.book
    worksheet = writer.sheets[sheet_name]
 
    # Get the dimensions of the dataframe.
    
    (max_row, max_col) = join_RTTA.shape   
    
    # Create a list of column headers, to use in add_table().
    
    column_settings = []
    for header in join_RTTA.columns:
        column_settings.append({'header': header})
    
    # Add the table.
    worksheet.add_table(2, 0, max_row, max_col - 1, {'columns': column_settings})

    #Add heading    
    worksheet.write(0, 0, 'RTTA Vulnerability Report '+datetime.now().strftime('%d %b %Y'), workbook.add_format({'bold': True, 'color': '#E26B0A', 'size': 15}))
        
    # Make the columns wider for clarity.
    worksheet.set_column(0, max_col - 1, 12)
    
###########FORMATTING Severity TABLE 1 #################### 

    worksheet = writer.sheets["Severity"]    
    (max_row, max_col) = Severity_table.shape
    column_settings = []
    for header in Severity_table.columns:
        column_settings.append({'header': header}) 
    worksheet.add_table(2, 0, max_row+2, max_col - 1, {'columns': column_settings})
    worksheet.write(0, 0, 'Vulnerabilities by Severity ', workbook.add_format({'bold': True, 'color': '#E26B0A', 'size': 15}))
    worksheet.set_column(0, max_col - 1, 30)
    
 ###########FORMATTING Severity TABLE 2 #################### 
    
    worksheet.write(max_row+5, 0, 'Severity 5 = Urgent ', workbook.add_format({'color': '#E26B0A', 'size': 12}))
    worksheet.write(max_row+6, 0, 'Severity 4 = Critical ', workbook.add_format({'color': '#E26B0A', 'size': 12}))
    worksheet.write(max_row+7, 0, 'Severity 3 = Serious ', workbook.add_format({'color': '#E26B0A', 'size': 12}))
    worksheet.write(max_row+8, 0, 'Severity 2 = Medium ', workbook.add_format({'color': '#E26B0A', 'size': 12}))
    worksheet.write(max_row+9, 0, 'Severity 1 = Minimal ', workbook.add_format({'color': '#E26B0A', 'size': 12}))
    
###########FORMATTING Asset TABLE #################### 

    worksheet = writer.sheets["Assets"]    
    (max_row, max_col) = asset_table.shape
    column_settings = []
    for header in asset_table.columns:
        column_settings.append({'header': header}) 
    worksheet.add_table(2, 0, max_row+2, max_col - 1, {'columns': column_settings})
    worksheet.write(0, 0, 'Vulnerabilities by Assets ', workbook.add_format({'bold': True, 'color': '#E26B0A', 'size': 15}))
    worksheet.set_column(0, max_col - 1, 30)
    
###########FORMATTING category TABLE #################### 

    worksheet = writer.sheets["Category"]    
    (max_row, max_col) = category_table.shape
    column_settings = []
    for header in category_table.columns:
        column_settings.append({'header': header}) 
    worksheet.add_table(2, 0, max_row+2, max_col - 1, {'columns': column_settings})
    worksheet.write(0, 0, 'Vulnerabilities by Category ', workbook.add_format({'bold': True, 'color': '#E26B0A', 'size': 15}))
    worksheet.set_column(0, max_col - 1, 30)
```
